```python
import numpy as np
import matplotlib.pyplot as plt

# Simular lanzamientos de una moneda justa (cara=1, cruz=0)
np.random.seed(42)  # Fijar semilla para reproducibilidad
n_experiments = 10000  # Número total de experimentos
lanzamientos = np.random.choice([0, 1], size=n_experiments)  # 0 = cruz, 1 = cara

# Calcular la frecuencia relativa acumulada de "cara"
frecuencia_relativa = np.cumsum(lanzamientos) / np.arange(1, n_experiments + 1)

# Visualizar cómo la frecuencia relativa converge a 0.5
plt.figure(figsize=(10, 6))
plt.plot(frecuencia_relativa, label="Frecuencia relativa acumulada (cara)")
plt.axhline(y=0.5, color="red", linestyle="--", label="Probabilidad verdadera (0.5)")
plt.title("Convergencia de la frecuencia relativa al aumentar n")
plt.xlabel("Número de experimentos (n)")
plt.ylabel("Frecuencia relativa de cara")
plt.legend()
plt.grid()
plt.show()
```
