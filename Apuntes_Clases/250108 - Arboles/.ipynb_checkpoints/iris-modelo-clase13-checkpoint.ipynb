```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns; sns.set()  # para mejorar la visualizaci√≥n

# Cargar el conjunto de datos de Iris
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
```

```python
# Definir colores para cada clase
colors = ['blue', 'green', 'red']

# An√°lisis exploratorio de datos: Histogramas de cada caracter√≠stica
fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax = ax.ravel()
for i in range(4):
    # Separar los datos por clase y hacer un histograma para cada uno
    for label, color in zip(range(len(target_names)), colors):
        ax[i].hist(X[y == label, i], bins=20, color=color, alpha=0.7, label=target_names[label])
    ax[i].set_title(f'Histograma de {feature_names[i]}')
    ax[i].set_xlabel('Valor')
    ax[i].set_ylabel('Frecuencia')
    ax[i].legend()

plt.tight_layout()
plt.show()
```

```python
# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear un modelo de √°rbol de decisi√≥n
tree = DecisionTreeClassifier(random_state=42)

# Entrenamiento del modelo
tree.fit(X_train, y_train)

# Visualizaci√≥n del √°rbol de decisi√≥n
plt.figure(figsize=(20,10))
plot_tree(tree, filled=True, feature_names=feature_names, class_names=list(target_names))
plt.title('Visualizaci√≥n del √Årbol de Decisi√≥n')
plt.show()
```

```python
# Evaluaci√≥n del modelo con datos de prueba
y_pred = tree.predict(X_test)

# Matriz de confusi√≥n
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Matriz de Confusi√≥n para el Clasificador de √Årboles de Decisi√≥n')
plt.show()

# Reporte de clasificaci√≥n
print("Evaluaci√≥n del modelo en el conjunto de prueba:")
print(classification_report(y_test, y_pred, target_names=target_names))

# Cross Validation para evaluar la estabilidad del modelo
scores = cross_val_score(tree, X_train, y_train, cv=5, scoring='accuracy')
print(f"Accuracy promedio de Cross Validation: {np.mean(scores):.2f} +/- {np.std(scores):.2f}")
```

```python
# Ajuste de hiperpar√°metros utilizando GridSearchCV
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': range(1, 10),
    'min_samples_split': range(2, 10),
    'min_samples_leaf': range(1, 10)
}
grid_search = GridSearchCV(tree, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Mejores par√°metros
print("Mejores par√°metros encontrados:")
print(grid_search.best_params_)

# Evaluaci√≥n con los mejores par√°metros
best_tree = grid_search.best_estimator_
y_pred_best = best_tree.predict(X_test)
print("Evaluaci√≥n del mejor modelo en el conjunto de prueba:")
print(classification_report(y_test, y_pred_best, target_names=target_names))
# Cross Validation para evaluar la estabilidad del modelo
print(f"Accuracy promedio de Cross Validation: {np.mean(scores):.2f} +/- {np.std(scores):.2f}")
# Graficar el √°rbol con los mejores par√°metros
plt.figure(figsize=(20,10))
plot_tree(best_tree, filled=True, feature_names=feature_names, class_names=list(target_names))
plt.title('Visualizaci√≥n del √Årbol de Decisi√≥n con Mejores Par√°metros')
plt.show()
```
