```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

1. Distribución Uniforme (PDF y CDF) 

Genera una distribución uniforme en el intervalo [0, 10]. Calcula y grafica la PDF y la CDF en ese rango. Indica el valor de la probabilidad acumulada para x=5

```python
# The a and b parameters.
a = 0
b = 10
base = b - a
st.uniform.pdf(5,a,base)

# Graficar la PDF y la CDF
plt.figure(figsize=(12, 6))

# Rango de valores para graficar
x = np.linspace(-6, 15, 10000)

# PDF
plt.subplot(1, 2, 1)
plt.plot(x, st.uniform.pdf(x,a,base), color='red', label='PDF')
plt.title('Función de Densidad de Probabilidad (PDF)')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.grid(True)
plt.legend()

#CDF
plt.subplot(1, 2, 2)
plt.plot(x, st.uniform.cdf(x, a, base), color='green', label='CDF')
plt.title('Función de Distribución Acumulada (CDF)')
plt.xlabel('x')
plt.ylabel('Probabilidad Acumulada')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

2. Distribución Gaussiana (PDF y PPF) 

Crea una distribución normal con media m=0 y desviación estándar σ=1. Grafica su PDF y calcula los percentiles 25, 50 y 75 utilizando la función PPF. 

```python
# Parámetros de la distribución
mu = 0
sigma = 1
x_num = 3
x = np.linspace(-x_num, x_num, 1000)
pdf = st.norm.pdf(x, mu, sigma)

p25 = st.norm.ppf(0.25, mu, sigma)
p50 = st.norm.ppf(0.50, mu, sigma)
p75 = st.norm.ppf(0.75, mu, sigma)

plt.figure(figsize=(14, 5))
plt.plot(x, pdf, label='PDF de la distribución normal')
plt.title('Distribución Normal (m=0, σ=1)')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)

plt.axvline(p25, color='r', linestyle='--', label=f'Percentil 25: {p25:.2f}')
plt.axvline(p50, color='g', linestyle='--', label=f'Percentil 50: {p50:.2f}')
plt.axvline(p75, color='b', linestyle='--', label=f'Percentil 75: {p75:.2f}')

plt.legend()
plt.show()

print(f"Percentil 25: {p25:.2f}")
print(f"Percentil 50 (mediana): {p50:.2f}")
print(f"Percentil 75: {p75:.2f}")
```

3. Distribución Chi-Cuadrado (PDF y CDF) 

Diseña una distribución chi-cuadrado con 5 grados de libertad. Grafica su PDF y determina la probabilidad acumulada para x=7 utilizando la CDF.

4. Comparación de distribuciones (CDF y PPF) 

Para una distribución uniforme en [0, 1], una normal estándar y una chi-cuadrado con 3 grados de libertad, compara las probabilidades acumuladas para x=0.5 en las tres distribuciones usando la CDF. 

```python
x_vals = np.linspace(0, 3, 1000)
x = 0.5
# CDF de cada distribución
uniform_cdf = st.uniform.cdf(x_vals, loc=0, scale=1)
normal_cdf = st.norm.cdf(x_vals)
chi2_cdf = st.chi2.cdf(x_vals, df=3)
# CDF en x = 0.5
uniform_cdf_x = st.uniform.cdf(x, loc=0, scale=1)
normal_cdf_x = st.norm.cdf(x)
chi2_cdf_x = st.chi2.cdf(x, df=3)

# Plottear
plt.figure(figsize=(10, 5))
plt.plot(x_vals, uniform_cdf, label='Uniforme [0, 1]', color='blue')
plt.plot(x_vals, normal_cdf, label='Normal Estándar', color='green')
plt.plot(x_vals, chi2_cdf, label='Chi-cuadrado (df=3)', color='red')
plt.axvline(x, color='black', linestyle='--', label=f'X = {x}')

# Control de Anotaciones
plt.scatter(x, uniform_cdf_x, color='blue', marker='x', s=100)
plt.text(x + 0.05, uniform_cdf_x, f'{uniform_cdf_x:.2f}', color='blue')
plt.scatter(x, normal_cdf_x, color='green', marker='x', s=100)
plt.text(x + 0.05, normal_cdf_x, f'{normal_cdf_x:.2f}', color='green')
plt.scatter(x, chi2_cdf_x, color='red', marker='x', s=100)
plt.text(x + 0.05, chi2_cdf_x, f'{chi2_cdf_x:.2f}', color='red')

plt.title('Comparación de CDFs para x=0.5')
plt.xlabel('x')
plt.ylabel('Probabilidad Acumulada (CDF)')
plt.grid(True)
plt.legend()
plt.show()

print(f"Probabilidad acumulada para x={x}:")
print(f"- Uniforme [0, 1]: {uniform_cdf_x}")
print(f"- Normal estándar: {normal_cdf_x}")
print(f"- Chi-cuadrado (df=3): {chi2_cdf_x}")
```

5. Probabilidades inversas (PPF) 

Para una distribución normal estándar, calcula los valores de x  que corresponden a probabilidades acumuladas de 0.1, 0.5 y 0.9 utilizando la PPF. 

```python
# Probabilidades acumuladas
probabilidades = [0.1, 0.5, 0.9]

# Cálculo de los valores de x usando PPF
valores_x = [st.norm.ppf(p) for p in probabilidades]

# Imprimir los valores de x
for p, x in zip(probabilidades, valores_x):
    print(f"Para una probabilidad acumulada de {p}, el valor de x es: {x:.2f}")

# Distribución normal estándar
x_vals = np.linspace(-3, 3, 1000)
normal_pdf = st.norm.pdf(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, normal_pdf, label='Normal Estándar (PDF)', color='blue')

for x in valores_x:
    plt.axvline(x, color='black', linestyle='--', linewidth=1)

plt.title('Distribución Normal Estándar y Líneas PPF')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad (PDF)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

6. Ajuste a la distribución uniformemente probable (PDF y CDF) 

Genera una distribución uniforme en [5, 15]. Encuentra la probabilidad de que un valor generado aleatoriamente esté entre 8 y 12 utilizando la CDF

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

a=5
b=15
x = np.linspace(a - 2, b + 2, 1000)

# Calcular la PDF y CDF
uniform_pdf = st.uniform.pdf(x, loc=a, scale=b - a)
uniform_cdf = st.uniform.cdf(x, loc=a, scale=b - a)

# Calcular la probabilidad de que el valor esté entre 8 y 12 usando la CDF
p_8 = st.uniform.cdf(8, loc=a, scale=b - a)
p_12 = st.uniform.cdf(12, loc=a, scale=b - a)
probabilidad_entre_8_y_12 = p_12 - p_8

# PDF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, uniform_pdf, color='blue', label='PDF')
plt.fill_between(x, uniform_pdf, where=(x >= 8) & (x <= 12), color='skyblue', alpha=0.5, label='Área entre 8 y 12')
plt.title('Distribución Uniforme [5, 15] - PDF')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()

# CDF
plt.subplot(1, 2, 2)
plt.plot(x, uniform_cdf, color='green', label='CDF')
plt.axvline(8, color='red', linestyle='--', label='x = 8')
plt.axvline(12, color='red', linestyle='--', label='x = 12')
plt.title('Distribución Uniforme [5, 15] - CDF')
plt.xlabel('x')
plt.ylabel('Probabilidad Acumulada')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

print(f"La probabilidad de que el valor esté entre 8 y 12 es: {probabilidad_entre_8_y_12:.2f}")
```

7. Cálculo de valores extremos (PPF) 

En una distribución chi-cuadrado con 10 grados de libertad, calcula el valor de x que corresponde al 95% del área acumulada utilizando la PPF.

```python
# The degree of freedom.
df = 10
probabilidad_acumulada = 0.95

x_95 = st.chi2.ppf(probabilidad_acumulada, df=df)

x = np.linspace(0, x_95 + 5, 1000)
chi2_pdf = st.chi2.pdf(x, df=df)

plt.figure(figsize=(10, 6))
plt.plot(x, chi2_pdf, label=f'Chi-cuadrado (df={df})', color='blue')
plt.axvline(x_95, color='red', linestyle='--', label=f'PPF (95%): x = {x_95:.2f}')
plt.fill_between(x, chi2_pdf, where=(x >= x_95), color='skyblue', alpha=0.5, label='Área > 95%')
plt.title('Distribución Chi-cuadrado (df=10) y Valor Extremo (PPF)')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad (PDF)')
plt.grid(True)
plt.legend()
plt.show()
print(f"El valor de x que corresponde al 95% del área acumulada es: {x_95:.2f}")
```

8. Validación del Teorema Central del Límite (PDF) 

Simula la suma de 10 variables aleatorias generadas uniformemente en [0, 1]. Grafica la PDF de la suma obtenida y compara su forma con una distribución normal estándar.

```python
n_lanzamientos = 10  # Número de variables aleatorias uniformes
n_experimentos = 1000  # Número de experimentos

# Simulación: Suma de variables uniformes [0,1]
resultados = np.random.uniform(0, 1, size=(n_experimentos, n_lanzamientos))  # Matriz de resultados
suma_muestral = resultados.sum(axis=1)  # Suma de cada experimento

# Calcular estadísticas clave
media_global = np.mean(suma_muestral)
desviacion_global = np.std(suma_muestral)

x_vals = np.linspace(min(suma_muestral), max(suma_muestral), 1000)
pdf_normal = st.norm.pdf(x_vals, loc=media_global, scale=desviacion_global) # PDF

plt.figure(figsize=(10, 6))
plt.hist(suma_muestral, bins=30, density=True, alpha=0.6, color='skyblue', label='Suma de variables uniformes')
plt.plot(x_vals, pdf_normal, color='red', label=f'Normal Estándar (Media={media_global:.2f}, $\sigma$={desviacion_global:.2f})')
plt.title('Teorema Central del Límite (PDF)')
plt.xlabel('Suma Muestral')
plt.ylabel('Densidad de Probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

9. Intervalo de confianza (PPF y CDF) 

Encuentra el intervalo de confianza del 95% para una distribución gaussiana con media m=10 y desviación estándar σ=2, utilizando la PPF. 

```python
media = 10
desviacion = 2

limite_inferior = st.norm.ppf(0.025, loc=media, scale=desviacion)
limite_superior = st.norm.ppf(0.975, loc=media, scale=desviacion)

x_vals = np.linspace(media - 4 * desviacion, media + 4 * desviacion, 1000)
normal_cdf = st.norm.cdf(x_vals, loc=media, scale=desviacion)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, normal_cdf, label='Distribución Gaussiana (CDF)', color='blue')
plt.axvline(limite_inferior, color='red', linestyle='--', label=f'Límite Inferior: {limite_inferior:.2f}')
plt.axvline(limite_superior, color='green', linestyle='--', label=f'Límite Superior: {limite_superior:.2f}')
plt.fill_between(x_vals, normal_cdf, where=(x_vals >= limite_inferior) & (x_vals <= limite_superior), color='skyblue', alpha=0.3, label='Intervalo de Confianza 95%')
plt.title('Intervalo de confianza (PPF y CDF)')
plt.xlabel('x')
plt.ylabel('Probabilidad Acumulada')
plt.grid(True)
plt.legend()
plt.show()

print(f"El intervalo de confianza del 95% es: [{limite_inferior:.2f}, {limite_superior:.2f}]")
```

10. Probabilidad acumulada en diferentes distribuciones (CDF) 

Calcula y compara la probabilidad acumulada hasta x=2 para una distribución uniforme en [0, 3], una gaussiana con m=0 y σ=1, y una chi-cuadrado con 4 grados de libertad.

```python
x = 2

uniform_cdf = st.uniform.cdf(x, loc=0, scale=3)  # Uniforme [0, 3]
normal_cdf = st.norm.cdf(x, loc=0, scale=1)      # Gaussiana (m=0, σ=1)
chi2_cdf = st.chi2.cdf(x, df=4)                  # Chi-cuadrado con 4 grados de libertad

x_vals = np.linspace(0, 5, 1000)
uniform_cdf_vals = st.uniform.cdf(x_vals, loc=0, scale=3)
normal_cdf_vals = st.norm.cdf(x_vals, loc=0, scale=1)
chi2_cdf_vals = st.chi2.cdf(x_vals, df=4)

# CDF
plt.figure(figsize=(10, 6))
plt.plot(x_vals, uniform_cdf_vals, label='Uniforme [0, 3]', color='blue')
plt.plot(x_vals, normal_cdf_vals, label='Normal (m=0, σ=1)', color='green')
plt.plot(x_vals, chi2_cdf_vals, label='Chi-cuadrado (df=4)', color='red')
plt.axvline(x, color='gray', linestyle='--')
plt.title('Probabilidad acumulada hasta x=2 en diferentes distribuciones (CDF)')
plt.xlabel('x')
plt.ylabel('Probabilidad Acumulada (CDF)')
plt.grid(True)
plt.legend()
plt.show()

print(f"Probabilidad acumulada hasta x={x}:")
print(f"  Uniforme [0, 3]: {uniform_cdf:.2f}")
print(f"  Normal (m=0, σ=1): {normal_cdf:.2f}")
print(f"  Chi-cuadrado (df=4): {chi2_cdf:.2f}")
```
