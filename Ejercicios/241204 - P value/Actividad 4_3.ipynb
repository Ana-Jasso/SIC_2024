```python
from scipy.stats import ttest_ind  # Prueba t para dos muestras independientes
from scipy.stats import ttest_rel  # Prueba t para muestras pareadas
from scipy.stats import f_oneway   # ANOVA de un factor para comparar las medias de tres o más grupos independientes
from scipy.stats import chisquare  # Prueba de Chi-cuadrada de bondad de ajuste para comparar frecuencias observadas con las esperadas
```

Problema: Un investigador quiere comparar la eficacia de dos tratamientos médicos, A y B, para una enfermedad. Se reclutaron dos grupos de pacientes, y cada grupo recibió uno de los tratamientos. Se midió la mejora en los síntomas después de 4 semanas para cada paciente.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Estableciendo la semilla para reproducibilidad
np.random.seed(0)

# Simulando datos de presión arterial
# Suponemos que las reducciones en la presión arterial son medidas en mmHg
data_A = np.random.normal(loc=8, scale=2.0, size=50)  # Reducción promedio con tratamiento A
data_B = np.random.normal(loc=7.8, scale=2.0, size=50)  # Reducción promedio con tratamiento B

# Realizando la prueba t para muestras independientes
t_stat, p_value = stats.ttest_ind(data_A, data_B)

print(f"Estadístico t: {t_stat:.3f}, Valor p: {p_value:.3f}")

# Interpretación de los resultados
alfa = 0.05
if p_value < alfa:
    print("Hay diferencias significativas entre los grupos.")
else:
    print("No hay diferencias significativas entre los grupos.")

# Visualización de los datos
plt.figure(figsize=(8, 5))
plt.hist(data_A, color='blue', alpha=0.7, bins=15, label='Tratamiento A - Mayor reducción')
plt.hist(data_B, color='red', alpha=0.7, bins=15, label='Tratamiento B - Menor reducción')
plt.title('Comparación de la reducción de la presión arterial entre dos tratamientos')
plt.xlabel('Reducción de la presión arterial (mmHg)')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
# Prueba t de Student para muestras pareadas
np.random.seed(0)

# Datos simulados realistas de rendimiento
before_training = np.random.normal(loc=50, scale=10, size=30)
after_training = before_training + np.random.normal(loc=5, scale=5, size=30)  # Mejora significativa

# Realizando la prueba t de Student para muestras pareadas
t_stat, p_value = stats.ttest_ind(before_training, after_training)

print(f"Estadístico t: {t_stat:f}, Valor p: {p_value:f}")

# Interpretación de los resultados
alfa = 0.05
if p_value < alfa:
    print("Hay diferencias significativas entre los grupos.")
else:
    print("No hay diferencias significativas entre los grupos.")

# Gráfico
plt.figure(figsize=(8, 5))
plt.plot(before_training, label='Antes de la capacitación', color='blue')
plt.plot(after_training, label='Después de la capacitación', color='green')
plt.title('Rendimiento antes y después de la capacitación')
plt.ylabel('Puntuación de Rendimiento')
plt.legend()
plt.show()
```

```python
# ANOVA de un factor para comparar las medias de tres o más grupos independientes
np.random.seed(0)

# Simulando datos realistas de preferencias
young = np.random.normal(loc=8, scale=2, size=40)  # Los jóvenes pueden tener una preferencia más alta
adult = np.random.normal(loc=7.5, scale=2, size=40)
elder = np.random.normal(loc=6, scale=2, size=40)  # Los mayores pueden tener menos preferencia

# Prueba ANOVA
f_statistic, p_value = stats.f_oneway(young, adult, elder)

# Resultados de la prueba ANOVA
print(f'Estadístico F: {f_statistic:f}')
print(f'Valor p: {p_value:f}')

# Interpretación de los resultados
alfa = 0.05
if p_value < alfa:
    print("Hay diferencias significativas entre los grupos.")
else:
    print("No hay diferencias significativas entre los grupos.")

# Gráfico
plt.figure(figsize=(8, 5))
plt.boxplot([young, adult, elder], labels=['Jóvenes', 'Adultos', 'Mayores'])
plt.title('Preferencias de consumidores por grupo de edad')
plt.ylabel('Nivel de Preferencia')
plt.show()
```

```python
# ANOVA de un factor para comparar las medias de tres o más grupos independientes
np.random.seed(0)

machine1 = np.random.normal(loc=50, scale=0.5, size=40)
machine2 = np.random.normal(loc=50.5, scale=0.5, size=40)
machine3 = np.random.normal(loc=49.5, scale=0.5, size=40)

# Prueba ANOVA
f_statistic, p_value = stats.f_oneway(machine1, machine2, machine3)

# Resultados de la prueba ANOVA
print(f'Estadístico F: {f_statistic:f}')
print(f'Valor p: {p_value:f}')

# Interpretación de los resultados
alfa = 0.05
if p_value < alfa:
    print("Hay diferencias significativas entre los grupos.")
else:
    print("No hay diferencias significativas entre los grupos.")

# Gráfico
plt.figure(figsize=(8, 5))
plt.boxplot([machine1, machine2, machine3], labels=['Máquina 1', 'Máquina 2', 'Máquina 3'])
plt.title('Peso de tornillos por máquina')
plt.ylabel('Peso')
plt.show()
```

```python
# Prueba de Chi-cuadrada de bondad de ajuste para comparar frecuencias observadas con las esperadas
np.random.seed(0)

# Simulando los datos
observed_fish_counts = np.array([10, 20, 30, 40, 50])
expected_fish_counts = np.array([30, 30, 30, 30, 30])

# Prueba de chi-cuadrado
chi2_statistic, p_value = stats.chisquare(observed_fish_counts,expected_fish_counts)

#Resultados
print(f'Estadístico F: {f_statistic:f}')
print(f'Valor p: {p_value:f}')

# Interpretar los resultados
alpha = 0.05
if p_value < alpha:
    print("Hay una diferencia significativa entre la distribución observada y esperada")
else:
    print("No hay evidencia suficiente para rechazar la distribución esperada")

# Gráfico
plt.figure(figsize=(8, 5))
plt.bar(['Área 1', 'Área 2', 'Área 3', 'Área 4', 'Área 5'], observed_fish_counts, color='blue', label='Observado')
plt.axhline(y=30, color='red', linestyle='--', label='Esperado')
plt.title('Distribución de peces en diferentes áreas del lago')
plt.ylabel('Número de peces')
plt.legend()
plt.show()
```
