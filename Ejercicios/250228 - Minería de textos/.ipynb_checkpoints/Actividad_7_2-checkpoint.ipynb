```python
import requests
from bs4 import BeautifulSoup
from collections import Counter
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from urllib.parse import urljoin
from wordcloud import WordCloud
from sklearn.preprocessing import MultiLabelBinarizer
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
# Obtener y limpiar el texto
def get_text_from_url(url):
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Error al acceder a {url}")
        return ""
    
    soup = BeautifulSoup(response.text, 'html.parser')
    content = soup.find('div', {'id': 'mw-content-text'})
    if not content:
        return ""
    
    for element in content(['script', 'style', 'table', 'sup']):
        element.decompose()
    
    text = ' '.join(p.get_text() for p in content.find_all('p')).lower()
    text = re.sub(r'[^a-z\s]', '', text)  # Eliminar puntuación
    return text

# Eliminar palabras de parada y tokenizar
stop_words = set(stopwords.words('english'))

def clean_and_tokenize(text):
    tokens = text.split()
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

# Análisis de frecuencia
def analyze_word_frequency(tokens, num):
    return Counter(tokens).most_common(num)

# Crear nube de palabras
def create_wordcloud(tokens):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(tokens))
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Codificación One-Hot
def one_hot_encode(tokens):
    mlb = MultiLabelBinarizer()
    one_hot = mlb.fit_transform([tokens])
    print(pd.DataFrame(one_hot, columns=mlb.classes_))

def create_heatmap(tokens, num):
    top_words = [word for word, _ in analyze_word_frequency(tokens, num)]
    df = pd.DataFrame(0, index=top_words, columns=top_words)
    for i in range(len(tokens) - 1):
        if tokens[i] in top_words and tokens[i + 1] in top_words:
            df.loc[tokens[i], tokens[i + 1]] += 1
            df.loc[tokens[i + 1], tokens[i]] += 1  # Modo espejo
    plt.figure(figsize=(12, 8))
    sns.heatmap(df, annot=True, cmap="viridis")
    plt.title("Mapa de Calor de Concurrencia de Palabras")
    plt.show()

# Crear gráfico de red
def create_network_graph(tokens, num):
    top_words = [word for word, _ in analyze_word_frequency(tokens, num)]
    G = nx.Graph()
    G.add_nodes_from(top_words)
    for i in range(len(tokens) - 1):
        if tokens[i] in top_words and tokens[i + 1] in top_words:
            if G.has_edge(tokens[i], tokens[i + 1]):
                G[tokens[i]][tokens[i + 1]]['weight'] += 1
            else:
                G.add_edge(tokens[i], tokens[i + 1], weight=1)
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=500, font_size=10, edge_color='skyblue')
    plt.title("Gráfico de Red de Concurrencia de Palabras")
    plt.show()
```

# Main

```python
url = "https://en.wikipedia.org/wiki/Machine_learning"
text = get_text_from_url(url)
tokens = clean_and_tokenize(text)
```

## Frecuencia de palabras:

```python
analyze_word_frequency(tokens, 15)
```

## Nube de Palabras

```python
create_wordcloud(tokens)
```

## Codificación One-Hot

```python
one_hot_encode(tokens)
```

```python
create_heatmap(tokens, 20)
```

```python
create_network_graph(tokens, 20)
```
