# Análisis y Clasificación del Conjunto de Datos de Vinos
**Contexto:** Eres un científico de datos trabajando para una consultoría enológica. Se te ha proporcionado un conjunto de datos que contiene los resultados de un análisis químico de vinos cultivados en la misma región de Italia pero originarios de tres diferentes cultivares. El objetivo es desarrollar un modelo predictivo que pueda clasificar los vinos según el cultivar basándose en su composición química.

**Datos:** El conjunto de datos de vino incluye las siguientes características químicas: Alcohol, Ácido málico, Cenizas, Alcalinidad de las cenizas, Magnesio, Fenoles totales, Flavonoides, Fenoles no flavonoides, Proantocianidinas, Intensidad del color, Matiz, OD280/OD315 de vinos diluidos, y Prolina. Estas medidas están asociadas a 178 muestras de vino pertenecientes a tres diferentes clases.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Cargar el conjunto de datos de vino
wine = load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names
target_names = wine.target_names
```

```python
X
```

```python
y
```

```python
data = pd.DataFrame(X, columns=feature_names)
data['target'] = y
```

### 1. Análisis Exploratorio de Datos:

Realiza un análisis exploratorio inicial visualizando la distribución de cada característica química mediante histogramas. Asegúrate de que cada histograma use un color diferente para cada tipo de vino (clase).

```python
num_features = len(feature_names)
for i in range(0, num_features, 2):
    plt.figure(figsize=(16, 6))
    for j in range(2):
        if i + j < num_features:
            plt.subplot(1, 2, j + 1)
            sns.histplot(data=data, x=feature_names[i + j], hue='target', kde=True, bins=30, palette="viridis")
            plt.title(f"Distribución de {feature_names[i + j]} por tipo de vino")
            plt.xlabel(feature_names[i + j])
            plt.ylabel("Frecuencia")
            plt.legend(title="Tipo de vino", labels=wine.target_names)
    plt.tight_layout()
    plt.show()
```

### 2. Preparación de los Datos:

Divide el conjunto de datos en un conjunto de entrenamiento y un conjunto de prueba. El conjunto de prueba debe representar el 30% del conjunto de datos original.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### 3. Desarrollo del Modelo:

Utiliza un modelo de árbol de decisión para clasificar los vinos según su cultivar. Entrena este modelo con el conjunto de entrenamiento.

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

### 4. Visualización del Modelo:

Visualiza el árbol de decisión entrenado para entender cómo el modelo está tomando decisiones basadas en las características químicas de los vinos.

```python
plt.figure(figsize=(20, 10))
plot_tree(model, feature_names=feature_names, class_names=list(wine.target_names), filled=True, rounded=True)
plt.title("Árbol de Decisión para Clasificación de Vinos")
plt.show()
```

### 5. Evaluación del Modelo:

Evalúa el rendimiento del modelo utilizando el conjunto de prueba y presenta un informe que incluya la precisión del modelo, la matriz de confusión y otras métricas de evaluación relevantes.

```python
# Evaluación del modelo con datos de prueba
y_pred = model.predict(X_test)

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples', xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Matriz de Confusión para el Clasificador de Árboles de Decisión')
plt.show()

# Reporte de clasificación
print("Evaluación del modelo en el conjunto de prueba:")
print(classification_report(y_test, y_pred, target_names=target_names))

# Cross Validation para evaluar la estabilidad del modelo
scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
print(f"Accuracy promedio de Cross Validation: {np.mean(scores):.2f} +/- {np.std(scores):.2f}")
```

### 6. Análisis de Multicolinealidad:

Calcula y visualiza la matriz de correlación para investigar si existe alguna multicolinealidad significativa entre las características. Discute cómo la multicolinealidad podría afectar la interpretación del modelo de árbol de decisión.

```python
correlation_matrix = data.iloc[:, :-1].corr()
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Matriz de Correlación entre Características")
plt.show()
```

Una correlación extremadamente alta entre dos variables diferentes podría indicar un problema en los datos y afectar la interpretación del modelo de árbol de decisión. Esto podría deberse a una dependencia directa entre dichas variables.

Una correlación baja o cercana a cero entre dos variables indica que no existe una relación lineal fuerte entre ellas.
