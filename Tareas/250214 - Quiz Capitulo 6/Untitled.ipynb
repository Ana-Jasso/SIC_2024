```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler

# Generar datos de muestra
X, labels_true = make_blobs(n_samples=300, n_features=2, centers=3, cluster_std=5, random_state=123)

# Normalizar los datos
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Configurar la figura para la visualización
plt.figure(figsize=(12, 4))

# Visualizar los datos originales
plt.subplot(1, 4, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels_true, cmap='viridis', alpha=0.7)
plt.title('Dataset Original')

# Aplicar K-Means con k=3
kmeans = KMeans(n_clusters=3, random_state=123)
kmeans_labels = kmeans.fit_predict(X)
plt.subplot(1, 4, 2)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', alpha=0.7)
plt.title('K-Means (k=3)')

# Aplicar Agrupamiento Aglomerativo con k=3
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(X)
plt.subplot(1, 4, 3)
plt.scatter(X[:, 0], X[:, 1], c=agglo_labels, cmap='viridis', alpha=0.7)
plt.title('Agglomerativo (k=3)')

# Aplicar DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
plt.subplot(1, 4, 4)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', alpha=0.7)
plt.title('DBSCAN')

# Ajustar el diseño de la figura
plt.tight_layout()
plt.show()
```
